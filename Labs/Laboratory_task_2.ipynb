```python
import numpy as np

# Input and target
x = np.array([1, 0, 1])
y = 1

# Hidden layer weights (3 inputs × 2 hidden units)
W1 = np.array([
    [0.2, -0.3],
    [0.4,  0.1],
    [-0.5, 0.2]
])

# Output layer weights (2 hidden units → 1 output)
W2 = np.array([-0.3, -0.2])

# Biases
theta = np.array([-0.4, 0.2, 0.1])  # [θ1, θ2, θ3]

# ---- Forward Pass ----

# Hidden layer net input
Z_hidden = np.dot(x, W1) + theta[:2]

# Activation (ReLU)
h = np.maximum(0, Z_hidden)

# Output layer net input
Z_output = np.dot(W2, h) + theta[2]

# Assuming linear activation at the output
y_pred = Z_output

# Compute squared error
error = 0.5 * (y - y_pred) ** 2

# ---- Results ----
print("Hidden layer net inputs (Z1, Z2):", Z_hidden)
print("Hidden activations (h1, h2):", h)
print("Predicted output (ŷ):", y_pred)
print("Error (E):", error)
```

| Step                   | Description               | Result                                                                                                  |
| ---------------------- | ------------------------- | ------------------------------------------------------------------------------------------------------- |
| **Hidden net inputs**  | (Z_1 = -0.7), (Z_2 = 0.1) | → These are the weighted sums before activation.                                                        |
| **Hidden activations** | (h_1 = 0), (h_2 = 0.1)    | → After applying ReLU, only the positive value (0.1) remains active.                                    |
| **Predicted output**   | (\hat{y} = 0.08)          | → The model’s estimated output for the given input.                                                     |
| **Error**              | (E = 0.4232)              | → Computed using (E = 0.5(y - \hat{y})^2); this measures how far the prediction is from the true label. |


The forward pass shows that only one hidden neuron was activated, indicating that the ReLU function filtered out negative inputs. The predicted output (0.08) is much lower than the actual value (1), resulting in a relatively high error of 0.4232. This suggests that the current weights and biases are not yet well-tuned, and further training or weight adjustment is needed to improve the model’s accuracy.
